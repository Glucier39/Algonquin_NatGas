````python
import yfinance as yf
import pandas as pd

ticker = "CL=F"   # WTI Crude Oil Futures

print(f"Downloading hourly data for {ticker}...\n")

# Years 2000–2025
periods = [(f"{y}-01-01", f"{y+1}-01-01") for y in range(2000, 2025)]

frames = []

for start, end in periods:
    print(f"Fetching {start} → {end}...")
    df = yf.download(
        ticker,
        start=start,
        end=end,
        interval="1h",
        progress=False
    )

    if df.empty:
        print(f"  ⚠️ No data returned for {start} → {end}")
        continue

    # Handle possible MultiIndex columns (e.g., ("Close","CL=F"))
    if isinstance(df.columns, pd.MultiIndex):
        df.columns = [c[0] for c in df.columns]

    df.reset_index(inplace=True)
    frames.append(df)

# Merge all non-empty chunks
if frames:
    wti = pd.concat(frames, ignore_index=True)
    wti = wti.drop_duplicates(subset="Datetime").sort_values("Datetime")

    # Rename columns for clarity
    wti.rename(columns={
        "Datetime": "Timestamp",
        "Open": "WTI_Open",
        "High": "WTI_High",
        "Low": "WTI_Low",
        "Close": "WTI_Close",
        "Volume": "WTI_Volume"
    }, inplace=True)
````

## Bloomberg Terminal API Setup

### Finding Bloomberg API on Terminal:
1. **Open Bloomberg Terminal** and log in
2. **Type `WAPI`** in the command line and press Enter
   - This opens the Bloomberg API Developer's Guide
   - Contains documentation, downloads, and sample code
3. **Navigate to "API Downloads"** section in WAPI
4. **Download Bloomberg API** for your operating system (Windows/macOS/Linux)

### Installation:
```bash
pip install blpapi
```

### Important Notes:
- Bloomberg Terminal must be **running and authenticated** on your machine
- The API connects to the local Bloomberg Terminal instance
- You must have Bloomberg Terminal installed (not just the API library)


````python
# Bloomberg Terminal API Example
# Make sure Bloomberg Terminal is running and you're logged in before running this

try:
    import blpapi
    import pandas as pd
    from datetime import datetime, timedelta
    
    # Create session and service
    session = blpapi.Session()
    if not session.start():
        print("Failed to start Bloomberg session. Make sure Terminal is running.")
    else:
        print("Bloomberg session started successfully!")
        
        # Open reference data service
        if not session.openService("//blp/refdata"):
            print("Failed to open //blp/refdata service")
        else:
            refDataService = session.getService("//blp/refdata")
            request = refDataService.createRequest("ReferenceDataRequest")
            
            # Example: Request data for a security (e.g., WTI Crude Oil)
            # Bloomberg ticker format: "CL1 COMDTY" for WTI Crude Oil front month
            request.append("securities", "CL1 COMDTY")  # WTI Crude Oil
            request.append("fields", "PX_LAST")  # Last price
            request.append("fields", "VOLUME")   # Volume
            
            # Send request
            session.sendRequest(request)
            
            # Process response
            while True:
                event = session.nextEvent(500)  # 500ms timeout
                if event.eventType() == blpapi.Event.RESPONSE or event.eventType() == blpapi.Event.PARTIAL_RESPONSE:
                    for msg in event:
                        print("Received:", msg)
                if event.eventType() == blpapi.Event.RESPONSE:
                    break
            
            session.stop()
            print("Bloomberg session closed.")

except ImportError:
    print("blpapi not installed. Install with: pip install blpapi")
except Exception as e:
    print(f"Error: {e}")
    print("Make sure Bloomberg Terminal is running and you're logged in.")
````

### Alternative: Using pdblp (Easier pandas integration)

`pdblp` is a wrapper around blpapi that makes it easier to work with pandas DataFrames.


````python
# Bloomberg API using pdblp (simpler pandas integration)
# Install with: pip install pdblp

try:
    import pdblp
    import pandas as pd
    
    # Connect to Bloomberg (Terminal must be running)
    con = pdblp.BCon(debug=False, port=8194, timeout=5000)
    con.start()
    
    # Example: Get current price for WTI Crude Oil
    # Bloomberg ticker: "CL1 COMDTY" (WTI Crude Oil front month)
    tickers = ["CL1 COMDTY"]  # Can add multiple tickers
    fields = ["PX_LAST", "VOLUME", "OPEN", "HIGH", "LOW"]  # Price fields
    
    # Get reference data
    df = con.bdp(tickers, fields)
    print("Current Bloomberg Data:")
    print(df)
    
    # Example: Get historical data
    # df_hist = con.bdh(
    #     tickers=["CL1 COMDTY"],
    #     flds=["PX_LAST", "VOLUME"],
    #     start_date="20240101",
    #     end_date="20241231"
    # )
    # print(df_hist)
    
    con.stop()
    
except ImportError:
    print("pdblp not installed. Install with: pip install pdblp")
    print("Note: You still need Bloomberg Terminal and blpapi installed")
except Exception as e:
    print(f"Error: {e}")
    print("Make sure Bloomberg Terminal is running and you're logged in.")
    print("Default port is 8194. If using different port, specify in BCon(port=XXXX)")
````
